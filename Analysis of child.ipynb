```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
data = pd.read_csv("child.csv")
```

```python
X = data[['age_months', 'weight_kg']]
y = data['label']
```

```python
model = LogisticRegression()
```

```python
model.fit(X, y)
```

```python
age_months = float(input("Enter the child's age in months(5-36): "))
weight_kg = float(input("Enter the child's weight in kilograms: "))
```

```python
user_input = pd.DataFrame({'age_months': [age_months], 'weight_kg': [weight_kg]})
```

```python
proba = model.predict_proba(user_input)
```

```python
custom_threshold = 0.5
```

```python
prediction = (proba[:, 1] >= custom_threshold).astype(int)
```

```python
if prediction[0] == 1:
    status = "healthy"
else:
    status = "unhealthy"

print(f"Based on the input, the child's status is predicted as: {status}")
```

```python
plt.scatter(data['age_months'], data['weight_kg'], c=data['label'], cmap='coolwarm', marker='o')
plt.xlabel("Age in Months")
plt.ylabel("Weight in Kilograms")
plt.title("Child Health Status Prediction")
plt.colorbar(label="Label (0: Unhealthy, 1: Healthy)")
plt.scatter(age_months, weight_kg, c=prediction, cmap='coolwarm', marker='X', s=100, label="Input")
plt.legend()
plt.show()
```
